# Long short Term Memory (LSTM)
image.png

Information is **added** or **removed** through structures called **gates**

Gates optionally let information through, for example via a sigmoid neural net layer and pointwise multiplication. 

# How do LSTMs word?
LSTMs are used to overcome the issue of **vanishing gradient** in RNNs.


1.   **Forget** irrelevant parts of the previous state
2.   **Store** relevant information
3.   **Update** the internal states *selectively*
4.   **Output**

## Key Concepts of LSTMs

1. Maintain a **separate cell state** from what is outputted
2. Use **gates** to control the **flow of information**
> 1. **Forget** gate gets rid of irrelevant information
> 2. **Store** relevant information from current input
> 3. Selectively **update** the cell state
> 4. **Output** gate return a filtered version of the cell state
3. Backpropagation through time with **uninterrupted gradient flow**
